```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# ChromeDriver Path
chrome_driver_path = r"C:\Users\DELL\Downloads\chromedriver-win64\chromedriver.exe"

# Set up WebDriver
service = Service(executable_path=chrome_driver_path)
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = webdriver.Chrome(service=service, options=options)

# Base URL with placeholders
base_url = "https://eic.mn/forestfire/index.php?item=5&count=100&page=1&action=more&fire_id={fire_id}"

# Range of fire_ids
start_fire_id = 10560
end_fire_id = 10569

# Log file to track results
output_file = "fire_id_check_results.csv"

# Open output file for writing
with open(output_file, "w", encoding="utf-8") as file:
    file.write("Fire_ID,Content_Available\n")

    for fire_id in range(start_fire_id, end_fire_id + 1):
        url = base_url.format(fire_id=fire_id)
        print(f"Checking URL: {url}")
        driver.get(url)
        time.sleep(2)  # Wait for the page to load

        try:
            # Check if specific content is available
            content = driver.find_element(By.TAG_NAME, "body").text
            if "Таны хайсан мэдээлэл байхгүй байна" in content:  # Adjust this string to match the page's "no content" message
                print(f"Fire ID {fire_id}: No content available")
                file.write(f"{fire_id},No\n")
            else:
                print(f"Fire ID {fire_id}: Content available")
                file.write(f"{fire_id},Yes\n")
        except Exception as e:
            print(f"Error checking Fire ID {fire_id}: {e}")
            file.write(f"{fire_id},Error\n")

# Close the driver
driver.quit()

print(f"Results saved to {output_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import os

# ChromeDriver Path
chrome_driver_path = r"C:\Users\DELL\Downloads\chromedriver-win64\chromedriver.exe"

# Output file path
output_path = r"C:\Users\DELL\Downloads\eic output\forest_fire_data.xlsx"

# Set up WebDriver
service = Service(executable_path=chrome_driver_path)
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = webdriver.Chrome(service=service, options=options)

# Base URL with placeholders
base_url = "https://eic.mn/forestfire/index.php?item=5&count=100&page=1&action=more&fire_id={fire_id}"

# Range of fire_ids
start_fire_id = 8892
end_fire_id = 10569

# Store the data
all_data = []

# Function to scrape the table
def scrape_fire_data(fire_id):
    url = base_url.format(fire_id=fire_id)
    print(f"Checking URL: {url}")
    driver.get(url)

    try:
        # Locate the table element
        table = driver.find_element(By.TAG_NAME, "table")
        rows = table.find_elements(By.TAG_NAME, "tr")

        fire_data = {"Fire_ID": fire_id}  # Start with Fire_ID column
        for row in rows:
            columns = row.find_elements(By.TAG_NAME, "th") + row.find_elements(By.TAG_NAME, "td")
            if len(columns) == 2:  # Ensure there are two columns
                key = columns[0].text.strip()  # Variable name
                value = columns[1].text.strip()  # Variable value
                fire_data[key] = value

        print(f"Fire ID {fire_id}: Data collected")
        return fire_data

    except Exception as e:
        print(f"Fire ID {fire_id}: No content or error occurred ({e})")
        return None

# Scrape data for all fire IDs
for fire_id in range(start_fire_id, end_fire_id + 1):
    fire_data = scrape_fire_data(fire_id)
    if fire_data:
        all_data.append(fire_data)

# Save to Excel
if all_data:
    # Create directory if not exists
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    # Convert to DataFrame and save
    df = pd.DataFrame(all_data)
    print(f"Columns in DataFrame: {df.columns}")  # Debug column structure
    df.to_excel(output_path, index=False, encoding="utf-8")
    print(f"Data saved to {output_path}")
else:
    print("No data available for the given range.")

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import os

# ChromeDriver Path
chrome_driver_path = r"C:\Users\DELL\Downloads\chromedriver-win64\chromedriver.exe"

# Output file path
output_path = r"C:\Users\DELL\Downloads\eic output\forest_fire_data.xlsx"

# Set up WebDriver
service = Service(executable_path=chrome_driver_path)
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option("useAutomationExtension", False)

driver = webdriver.Chrome(service=service, options=options)

# Base URL with placeholders
base_url = "https://eic.mn/forestfire/index.php?item=6&count=10&page=1&sort=3&sorttype=2&action=more&act_id={act_id}"

# Range of act_ids
start_act_id = 7754
end_act_id = 8788

# Store the data
all_data = []

# Function to scrape the table
def scrape_act_data(act_id):
    url = base_url.format(act_id=act_id)
    print(f"Checking URL: {url}")
    driver.get(url)

    try:
        # Locate the table element
        table = driver.find_element(By.TAG_NAME, "table")
        rows = table.find_elements(By.TAG_NAME, "tr")

        act_data = {"Act_ID": act_id}  # Start with Act_ID column
        for row in rows:
            columns = row.find_elements(By.TAG_NAME, "th") + row.find_elements(By.TAG_NAME, "td")
            if len(columns) == 2:  # Ensure there are two columns
                key = columns[0].text.strip()  # Variable name
                value = columns[1].text.strip()  # Variable value
                act_data[key] = value

        print(f"Act ID {act_id}: Data collected")
        return act_data

    except Exception as e:
        print(f"Act ID {act_id}: No content or error occurred ({e})")
        return None

# Scrape data for all act IDs
for act_id in range(start_act_id, end_act_id + 1):
    act_data = scrape_act_data(act_id)
    if act_data:
        all_data.append(act_data)

# Save to Excel in a new sheet
if all_data:
    # Create directory if not exists
    os.makedirs(os.path.dirname(output_path), exist_ok=True)

    # Load the existing Excel file or create a new one
    try:
        with pd.ExcelWriter(output_path, mode="a", if_sheet_exists="overlay", engine="openpyxl") as writer:
            df = pd.DataFrame(all_data)
            print(f"Columns in DataFrame: {df.columns}")  # Debug column structure
            df.to_excel(writer, sheet_name="Act_Data", index=False, encoding="utf-8")
            print(f"Data saved to new sheet in {output_path}")
    except FileNotFoundError:
        with pd.ExcelWriter(output_path, mode="w", engine="openpyxl") as writer:
            df = pd.DataFrame(all_data)
            print(f"Columns in DataFrame: {df.columns}")  # Debug column structure
            df.to_excel(writer, sheet_name="Act_Data", index=False, encoding="utf-8")
            print(f"Data saved to {output_path} (new file)")
else:
    print("No data available for the given range.")

# Close the driver
driver.quit()
```
